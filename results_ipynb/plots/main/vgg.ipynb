```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
from tang_jcompneuro.model_fitting_postprocess import load_data_generic
from tang_jcompneuro.model_fitting_cnnpre import blob_corresponding_info, subtype_to_train
```

```python
def modelname_alternative(_1, x, _2, _3):
    net_name, _, blob = x.split('+')
    assert _ == 'legacy'
    return net_name + '$' +',{:02d},'.format(list(blob_corresponding_info[net_name]).index(blob)) + blob
```

```python
def check_all(squared, score_col_name):
    models_to_examine = [
        ('cnnpre', x) for x in subtype_to_train  # both bn or nonbn
    ]
    
    
    return load_data_generic(models_to_examine, load_naive=False, metric='ccnorm_5', squared=squared,
                             percentages_to_load=(100,),
                            score_col_name=score_col_name, modelname_alternative=modelname_alternative)
```

```python
df_all_cc2 = check_all(squared=True, score_col_name='cc2').xs(100, level='percentage').sort_index()
```

```python
HO_neuron_perf = df_all_cc2.apply(lambda x: x['cc2']['HO']['mean'], axis=1).unstack('subset')
HO_neuron_perf
```

```python
OT_neuron_perf = df_all_cc2.apply(lambda x: x['cc2']['OT']['mean'], axis=1).unstack('subset')
OT_neuron_perf
```

```python
from tang_jcompneuro.plotting import image_subset_and_neuron_subset_list, show_perf_vs_size
from tang_jcompneuro import dir_dictionary
import os.path
```

```python
# time to plot.
# magic numbers here follow
# https://github.com/leelabcnbc/tang_jcompneuro/blob/master/thesis_plots/v1_fitting/results_performance_vgg.ipynb
def draw_one_stuff(dataset, net_name, letter):
    # draw one by one.
    
    num_panel = len(image_subset_and_neuron_subset_list)
    
    plt.close('all')
    fig, ax = plt.subplots(1, 1, squeeze=True, figsize=(5.5,2.5))
    
    color_list = ('#1f77b4', '#ff7f0e', '#2ca02c')
    data_x_first = None
    xdatas = []
    ydatas = []
    labels = []
    style_list = []
    for idx, (img_subset, neuron_subset) in enumerate(image_subset_and_neuron_subset_list):
            
        if neuron_subset == 'OT':
            data_y = OT_neuron_perf[img_subset].loc[dataset].values
            data_x = OT_neuron_perf[img_subset].loc[dataset].index.values
        elif neuron_subset == 'HO':
            data_y = HO_neuron_perf[img_subset].loc[dataset].values
            data_x = HO_neuron_perf[img_subset].loc[dataset].index.values
        else:
            raise NotImplementedError
        data_y = np.asarray([v for idx_v, v in enumerate(data_y) if data_x[idx_v].startswith(f'{net_name}$')])
        data_x = [v[len(net_name)+5:] for idx_v, v in enumerate(data_x) if data_x[idx_v].startswith(f'{net_name}$')]
#         print(data_y.shape, data_x)
        
        if data_x_first is None:
            data_x_first = data_x
        assert data_x_first == data_x
        
        xdatas.append(np.arange(len(data_x)))
        ydatas.append(data_y)
        labels.append(f'{neuron_subset} neurons\n{img_subset} stimuli')
        style_list.append(('-', color_list[idx], 'x'))
        

    show_perf_vs_size(xdatas, ydatas, labels,
#                       title=f'{neuron_subset} neurons\n{img_subset} stimuli',
                      ax=ax,
                      ylabel='mean $CC_\mathrm{norm}^2$',
                      style_list=style_list, show_legend=True,
                      xticks=np.arange(len(data_x_first)),
                      xticklabels=data_x_first,
                      legend_param=dict(loc='best', ncol=3),
                      xlim=(-0.5, len(data_x_first)-0.5), xlabel=f'{net_name} layer',
                      xlabel_param={'rotation': 'vertical'}
                     )
#         )
#     mk = {'MkA_Shape': 'A', 'MkE2_Shape': 'B'}[dataset]
#     fig.suptitle(f'{net_name} on monkey {mk}')
#     # adjust figure 
    fig.text(0, 1, letter,
                            horizontalalignment='left',
                            verticalalignment='top', fontweight='bold',
                            transform=fig.transFigure)
    fig.subplots_adjust(top=0.95, bottom=0.4, left=0.13, right=0.99, hspace=0.1, wspace=0.05)
#     fig.subplots_adjust(top=0.75, bottom=0.2, left=0.11, right=0.96, hspace=0.1, wspace=0.075)
    save_dir = os.path.join(dir_dictionary['plots'], 'main', 'vgg')
    os.makedirs(save_dir, exist_ok=True)
    fig.savefig(os.path.join(save_dir, f'{net_name}_{dataset}.pdf'), dpi=300)
    
    plt.show()
```

```python
draw_one_stuff('MkA_Shape', 'vgg16', 'e')
draw_one_stuff('MkA_Shape', 'vgg19', 'g')
draw_one_stuff('MkA_Shape', 'vgg16_bn', 'd')
draw_one_stuff('MkA_Shape', 'vgg19_bn', 'f')
```

```python
draw_one_stuff('MkE2_Shape', 'vgg16', 'e')
draw_one_stuff('MkE2_Shape', 'vgg19', 'g')
draw_one_stuff('MkE2_Shape', 'vgg16_bn', 'd')
draw_one_stuff('MkE2_Shape', 'vgg19_bn', 'f')
```
