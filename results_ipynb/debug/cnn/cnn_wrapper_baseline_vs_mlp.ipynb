```python
model_list = (
    'b.9',
    'mlp.4', 'mlp.4@0.05',
    'mlp.40',
    'mlp.60',
    'mlp.80',
    'mlp.100',
    'mlp.145','mlp.145@0.05'
)
```

```python
def do_one_model(model_name):
    suffix_this = cnn_suffix_fn(model_name)
    topdim_this = cnn_top_dim_fn(model_name)
    print(model_name, suffix_this, topdim_this)
    
    datasets = load_split_dataset('MkA_Shape', 'all', True, 553, suffix=suffix_this,
                                 top_dim=topdim_this)
    trainer = get_trainer(model_name, cudnn_enabled=True,
                          show_arch_config=True)
    trainer(datasets)
```

```python
for x in model_list:
    do_one_model(x)
```
